### 【問題1】乱数の作成

```python
import numpy as np
import matplotlib.pyplot as plt

# 平均
average_data = [-3,0]

# 共分散
covariance_matrix = [[1.0,0.8],
                     [0.8,1.0]]

# 500個の乱数を作成
np.random.seed(0) # 乱数を固定
random_data = np.random.multivariate_normal(average_data,covariance_matrix,500)

# shapeを確認
print(random_data.shape)
```

### 【問題2】散布図による可視化

```python
# x軸、y軸の数値を取り出す
x = random_data[:,0]
y = random_data[:,1]

# 散布図にプロット
plt.xlabel("x")
plt.ylabel("y")
plt.title("Scatter plot of random numbers")
plt.scatter(x,y)
```

### 【問題3】ヒストグラムによる可視化

```python
# リスト０番目の要素のヒストグラム
plt.xlabel("x")
plt.ylabel("y")
plt.title("Histogram of hist1")
hist1 = plt.hist(x,bins = 500, color = "blue")
plt.xlim(-6,6)
plt.show(hist1)

# リスト１番目の要素のヒストグラム
plt.xlabel("x")
plt.ylabel("y")
plt.title("Histogram of hist2")
hist2 = plt.hist(y,bins = 500, color = "red")
plt.xlim(-6,6)
plt.show(hist2)
```

### 【問題4】データの追加

```python
# 平均
average_data1 = [0,-3]

# 共分散
covariance_matrix1 = [[1.0,0.8],
                     [0.8,1.0]]


# 500個の乱数を作成
np.random.seed(0) # 乱数を固定
random_data1 = np.random.multivariate_normal(average_data1,covariance_matrix1,500)

# shapeを確認
print(random_data1.shape)

# random_dataとrandom_data1の散布図

# random_data1の軸、y軸の数値を取り出す

x1 = random_data1[:,0]
y1 = random_data1[:,1]

# 散布図にプロット
plt.xlabel("x")
plt.ylabel("y")
plt.title("Scatter plot of random numbers")
scatter0 = plt.scatter(x,y)
scatter1 = plt.scatter(x1,y1)
plt.legend((scatter0, scatter1), ("= 0", "= 1"), loc=4)
plt.show()
```

### 【問題5】データの結合

```python
# random_dataとrandom_data1 のデータを結合
all_random_data = np.concatenate((random_data, random_data1), axis = 0)

# shapeとリスト内容を確認
print(all_random_data.shape)
print(all_random_data)
```
