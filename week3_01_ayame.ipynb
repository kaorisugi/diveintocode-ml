・特徴量を一つずつ選び、箱ひげ図を使ってラベルごとのデータの分布を可視化する

```python
# 特徴量sepal length (cm)

def plot_by_label(data_name):

    # ラベルの値のリストを取得
    label_Species = df["Species"].unique()
    #print(label_Species)

    # ラベル別に行を取り出す
    label_0 = df[df["Species"].isin(["0"])]
    label_1 = df[df["Species"].isin(["1"])]
    label_2 = df[df["Species"].isin(["2"])]

    # ラベル別のsepal length (cm)列を取り出す
    data_label_0 = np.array(label_0[data_name])
    data_label_1 = np.array(label_1[data_name])
    data_label_2 = np.array(label_2[data_name])
    #print(sepal_label_0)
    #print(sepal_label_1)
    #print(sepal_label_2)

    # ラベルの名前のリストを用意
    target_names = ["setosa", "versicolor", "virginica"]

    # 箱ひげ図に描画
    plt.title("Distribution of '" + data_name + "' per label")
    plt.ylabel(data_name)
    plt.boxplot([data_label_0, data_label_1, data_label_2], labels = target_names)
    plt.show()

plot_by_label(data_name="sepal length (cm)")
```

・特徴量を一つずつ選び、バイオリン図を使ってラベルごとのデータの分布を可視化する

```python
# 特徴量sepal length (cm)

def violinplot_by_label(data_name):

    # ラベルの値のリストを取得
    label_Species = df["Species"].unique()
    #print(label_Species)

    # ラベル別に行を取り出す
    label_0 = df[df["Species"].isin(["0"])]
    label_1 = df[df["Species"].isin(["1"])]
    label_2 = df[df["Species"].isin(["2"])]

    # ラベル別のsepal length (cm)列を取り出す
    data_label_0 = np.array(label_0[data_name])
    data_label_1 = np.array(label_1[data_name])
    data_label_2 = np.array(label_2[data_name])
    #print(data_label_0)
    #print(data_label_1)
    #print(data_label_2)

    # ラベルの名前のリストを用意
    target_names = ["setosa", "versicolor", "virginica"]

    # バイオリン図に描画
    fig = plt.figure()
    ax = fig.add_subplot(1, 1, 1)
    plt.title("Distribution of '" + data_name + "' per label")
    ax.violinplot([data_label_0, data_label_1, data_label_2])
    ax.set_xticks([1, 2, 3])
    ax.set_xticklabels(target_names)
    ax.set_ylabel(data_name)
    plt.show()

violinplot_by_label(data_name="sepal length (cm)")
```

箱ひげ図とバイオリン図は似た図を表示します。これらはどう違い、それぞれどのような利点があるのかについて説明してください

箱ひげ図はデータの分布を「箱」と「ひげ」で表したグラフで、データがどのあたりの値に集中しているかといった分布の範囲を四分位で表し、それをひと目で捉えることができます。<br>バイオリン図は、箱ひげ図の両脇に90度回転させたカーネル密度グラフを付加した図で、箱ひげ図よりも多くの情報量を含んでいる点が優れていますが、知名度が低いためにバイオリンプロットを知らない人には読み方がわからないのが欠点です。

### 【問題7】特徴量間の関係の確認

・縦軸と横軸に特徴量を１つずつ取り、種類毎に色分けした散布図を作る

```python
# x=sepal length (cm)、y=sepal width (cm)

def scatter_by_label(data_name_x, data_name_y):

    # ラベルの値のリストを取得
    label_Species = df["Species"].unique()
    #print(label_Species)

    # ラベル別に行を取り出す
    label_0 = df[df["Species"].isin(["0"])]
    label_1 = df[df["Species"].isin(["1"])]
    label_2 = df[df["Species"].isin(["2"])]
    #print(label_0)

    # ラベル別のsepal length (cm)、sepal width (cm)列を取り出す
    data_label_0 = label_0.loc[:,[data_name_x, data_name_y]]
    data_label_1 = label_1.loc[:,[data_name_x, data_name_y]]
    data_label_2 = label_2.loc[:,[data_name_x, data_name_y]]
    #print(data_label_0.shape)
    #print(data_label_1)
    #print(data_label_2)

    # ラベルの名前のリストを用意
    target_names = ["setosa", "versicolor", "virginica"]

    # 散布図に描画
    fig = plt.figure()
    ax = fig.add_subplot(1, 1, 1)
    plt.title("Scatter plot by label")
    label0 = ax.scatter(data_label_0[data_name_x], data_label_0[data_name_y], color = "red")
    label1 = ax.scatter(data_label_1[data_name_x], data_label_1[data_name_y], color = "blue")
    label2 = ax.scatter(data_label_2[data_name_x], data_label_2[data_name_y], color = "green")
    ax.legend((label0, label1, label2), (target_names), loc=4)
    ax.set_xticklabels(data_name_x)
    ax.set_xlabel(data_name_x)
    ax.set_ylabel(data_name_y)
    plt.show()

data_name_x = "sepal length (cm)"
data_name_y = "sepal width (cm)"

scatter_by_label(data_name_x, data_name_y)
```

```python
colormap = plt.cm.RdBu
plt.title('Heat map of correlation coefficient matrix', y=1.05, size=15)
sns.heatmap(corr_mat,
            vmin=-1.0,
            vmax=1.0,
            center=0,
            annot=True, # True:格子の中に各要素の値を表示
            fmt='.1f',
            xticklabels=corr_mat.columns.values,
            yticklabels=corr_mat.columns.values
           )
plt.show()
```
